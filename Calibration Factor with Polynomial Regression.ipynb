# Load Cell Calibration Factor using Polinomial Regression

## Prepared Scale

For this project, you should prepare an object with certain weight as referrence. This object must have known weight with real scale. Choose the number you want to scale. Then pick the range number of calibration factor. Make sure the value is in the range of measured scale. 

For example, I use a 1 kg of rice which I have measured before with real scale. I choose the 1000 and the range of calibration factor is 100-1000. Then, proceed with the arduino code.

## Get the Measured Value

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
calFactor_Y = ([100,200,300,400,500,600,700,800,900,1000]) #Range of the calibration factor
measuredValue_X = ([4336,2168,1445,1084,867,722,619,542,482,434]) #Measured Value (1000 is in the range)
```

```python
plt.scatter(measuredValue_X,calFactor_Y)
plt.xlabel('Measured Value (gr)')
plt.ylabel('Calibration Factor')
plt.show()
```

## Get Polynomial Eq using Numpy Polyfit

```python
poly = np.polyfit(measuredValue_X,calFactor_Y,4) #Get the eq by polynomial regression 4 order
```

```python
x = np.linspace(300,4500,1000)
plt.scatter(measuredValue_X,calFactor_Y)
plt.plot(x,np.polyval(poly,x),'y')
plt.xlabel('Measured Value (gr)')
plt.ylabel('Calibration Factor')
plt.grid()
plt.show()
```

```python
print('The Polynomial Equation : y = ',poly[0],'x^4 +',poly[1],'x^3 +',poly[2],'x^2 +',poly[3],'x^1 +',poly[4])
```

## Get Calibration Factor of Real Value with Poly Eq

```python
rightCalFactor = 0
myRealValue = 1000
for i in range(len(poly)):
    rightCalFactor += poly[i]*(myRealValue**(4-i))
print('The Calibration Factor for 1000 gram is : ', rightCalFactor)
```

```python
print('The Calibration Factor 1 kg is : ', rightCalFactor*1000)
```
